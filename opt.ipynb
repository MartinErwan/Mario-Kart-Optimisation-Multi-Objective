Nous disposons de 44 personnages, 21 roues, 40 chassis et 14 deltaplane. <br>
Nous avons donc 44 x 21 x 40 x 14 = 517 440 combinaisons possibles.<br>
Étant donné que le nombre de possibilité est fini, on peut générer un ensemble d'observations comportant toutes les observations possibles.

On définie une observation comme étant un vecteur de dimension 6 qui pour une combinaison donnée comporte :
- La vitesse totale : La vitesse de pointe de notre véhicule.
- L'accelération totale : L'acceleration de notre véhicule (capacité à changer de vitesse +/- rapidement)
- Le poids total : Le poids du véhicule (utile lors de coalision avec les autres joueurs)
- La maniabilité totale : La maniabilité du véhicule.
- L'adhérence totale : L'adhérence du véhicule sur des terrains pluvieux, sablonneux ou enneigés.
- Le mini turbo totale : La capacité à sortir un mini turbo* +/- rapidement après un dérapage.

<u>Mini turbo</u> : Dans Mario kart, lorsque vous effectuez un dérapage (notamment durant un virage) et que vous le maintenez assez longtemps le jeu vous donne un bonus de vitesse que l'on appelle 'Mini turbo'.
<div style="text-align:center">
    <img src="./img/img6.jpg" alt="Texte alternatif" width="500" height="300">
</div>

```python
#Génération d'une observation (= d'un individu)
def generate_indiv(perso,roue,kart,delta,df_perso,df_roues,df_kart,df_delta):
    Xi = {}
    Xi["Combos"] = "{},{},{},{}".format(perso,roue,kart,delta)
    
    Xi["Vitesse"] = float(df_perso[perso][0])+float(df_delta[delta][0])+float(df_kart[kart][0])+float(df_roues[roue][0])
    Xi["Acceleration"] = float(df_perso[perso][1])+float(df_delta[delta][1])+float(df_kart[kart][1])+float(df_roues[roue][1])
    Xi["Poids"] = float(df_perso[perso][2])+float(df_delta[delta][2])+float(df_kart[kart][2])+float(df_roues[roue][2])
    Xi["Maniabilité"] = float(df_perso[perso][3])+float(df_delta[delta][3])+float(df_kart[kart][3])+float(df_roues[roue][3])
    Xi["Adhérence"] = float(df_perso[perso][4])+float(df_delta[delta][4])+float(df_kart[kart][4])+float(df_roues[roue][4])
    Xi["Mini_Turbo"] = float(df_perso[perso][5])+float(df_delta[delta][5])+float(df_kart[kart][5])+float(df_roues[roue][5])
    return Xi 

#Génération d'une population (= d'un ensemble d'individus)
def generate_pop(df_perso,df_roues,df_kart,df_delta):

    personnages = df_perso.columns
    roues = df_roues.columns
    karts = df_kart.columns
    deltas = df_delta.columns
    X = [] #Liste qui contiendra toutes les combinaisons possibles
    #Pour chaque combinaison (perso,roue,kart,delta) possible, X apprend le vecteur Xi correspondant
    for perso in personnages:
        for roue in roues:
            for kart in karts:
                for delta in deltas:
                    Xi = generate_indiv(perso,roue,kart,delta,df_perso,df_roues,df_kart,df_delta)
                    X.append(Xi)
                    
    return X
    
```

```python
X = generate_pop(df_perso,df_roues,df_kart,df_delta)
```

```python
DF = pd.DataFrame(X)
```

On va à présent s'intérésser aux points qui se situent sur la <b>frontière de Pareto</b>. Nous retrouvons sur cette frontière l'ensemble des optimas de Pareto. C'est-à-dire que sur cette frontière, il n'existe objectivement aucune solution meilleur que l'autre. On parle de solutions non dominées.<br>

<u>Exemple</u> : une solution s1 offre 5.5 en vitesse et 2.75 en mini turbo, une solution s2 offre 5.75 en vitesse et 2.25 en mini turbo. <br>

s1 offre un meilleur mini turbo que s2 mais s2 offre une meilleure vitesse que s1. Comment choisir?<br> 

Il faudrait faire une étude plus poussée en jeu pour savoir qu'est-ce qui est le plus important entre la vitesse le mini turbo et à quel point.<br> 

Pour l'instant, ce que je peux dire, d'après mes recherches, c'est que cela va grandement dépendre du choix de la piste. Sur une piste avec beaucoup de virages, il sera préférable de favoriser la statistique de mini turbo tandis que sur une piste avec beaucoup de ligne droite, la vitesse de pointe sera plus importante.

```python
def pareto_frontier(df, objective1, objective2, label_column):
    # On trie le dataframe en fonction du 1er objectif (ici la vitesse)
    sorted_df = df.sort_values(by=objective1)

    # On initialise la frontière de Pareto avec le premier point (qui n'est pas un point dominé grâce au tri)
    frontier = [(sorted_df.iloc[0][objective1], sorted_df.iloc[0][objective2], sorted_df.iloc[0][label_column],sorted_df.index[0])]
    # On regarde pour chaque point s'il est dominé ou non
    for index, row in sorted_df.iterrows():
        current_point = (row[objective1], row[objective2], row[label_column],index)
        is_dominated = False

        for point in frontier:
            # On vérifie si le point actuel domine l'un des points de la frontière de Pareto actuelle
            if (current_point[0] >= point[0]) and (current_point[1] >= point[1]):
                # Si c'est le cas, on retire le point de la frontière
                frontier.remove(point)

            # On vérifie si le point actuel est dominé par l'un des points de la frontière
            elif (current_point[0] <= point[0]) and (current_point[1] <= point[1]):
                #Si c'est le cas, on le tag comme étant dominé
                is_dominated = True
                break

        # S'il n'est pas dominé, on l'ajoute à la frontière
        if not is_dominated:
            frontier.append(current_point)

    return frontier


```

```python
frontier_points = pareto_frontier(DF, 'Vitesse', 'Mini_Turbo', 'Combos')
frontier_points
```

```python
indexes = [item[-1] for item in frontier_points]
```

```python
DF['Pareto'] = 0
DF.loc[indexes,'Pareto'] = 1
```

```python
DF[DF['Pareto']==1]
```

```python
for index in indexes:
    speed = DF.loc[index,'Vitesse']
    turbo = DF.loc[index,'Mini_Turbo']
    DF.loc[(DF['Vitesse']==speed) & (DF['Mini_Turbo']==turbo),'Pareto']=1
```

```python
fig = px.scatter(DF, x="Vitesse", y="Mini_Turbo",hover_name="Combos",color='Pareto',color_continuous_scale=['blue', 'red'],title="Vitesse et Mini Turbo de l'ensemble des combinaisons possibles")
fig.show()
```
